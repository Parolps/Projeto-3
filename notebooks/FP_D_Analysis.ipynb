```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import pickle
import seaborn as sns

pd.options.display.max_columns = None
plt.style.use("fivethirtyeight")
plt.rcParams["figure.dpi"] = 150
```

```python
X_train, X_ivs, y_train, col_names = pickle.load(open("../data/drd2_data.pickle", "rb"))
df = pd.DataFrame(X_train, columns=col_names)
df["target"] = y_train
df.head()
```

```python
FP_cols = [col for col in df.columns if "F" in col]
D_cols = [col for col in df.columns if "D" in col]
D_data = df[D_cols].copy()
FP_data = df[FP_cols].copy()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA

# scaler = StandardScaler()
scaler = MinMaxScaler()
D_data_scaled = scaler.fit_transform(D_data)


pca = PCA(n_components=2)

D_data_pca = pca.fit_transform(D_data_scaled)
D_data_pca = pd.DataFrame(D_data_pca, columns=["PC1", "PC2"])

D_data_pca["target"] = df["target"].values

plt.figure(figsize=(8, 6))
plt.scatter(
    x=D_data_pca["PC1"],
    y=D_data_pca["PC2"],
    c=D_data_pca["target"],
    cmap="viridis",
    alpha=0.5,
    edgecolors="k",
)
plt.title("PCA of D descriptors")
plt.show()
```

```python
# scaler = StandardScaler()
scaler = MinMaxScaler()
FP_data_scaled = scaler.fit_transform(FP_data)

pca = PCA(n_components=2)

FP_data_pca = pca.fit_transform(FP_data_scaled)
FP_data_pca = pd.DataFrame(FP_data_pca, columns=["PC1", "PC2"])
FP_data_pca["target"] = df["target"].values

plt.figure(figsize=(8, 6))
plt.scatter(
    x=FP_data_pca["PC1"],
    y=FP_data_pca["PC2"],
    c=FP_data_pca["target"],
    cmap="viridis",
    alpha=0.5,
    edgecolors="k",
)
plt.title("PCA of D descriptors")
plt.show()
```
