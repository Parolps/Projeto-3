# Vanilla models

```python
# regression
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import cross_validate
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import (
    RandomForestRegressor,
    GradientBoostingRegressor,
    AdaBoostRegressor,
)
from sklearn.metrics import (
    mean_squared_error,
    make_scorer,
    mean_absolute_error,
    median_absolute_error,
    explained_variance_score,
    max_error,
)

metrics_reg = {
    "MSE": make_scorer(mean_squared_error),
    "MAE": make_scorer(mean_absolute_error),
    "MAX_ERROR": make_scorer(max_error),
    "EV": make_scorer(explained_variance_score),
    "RMSE": make_scorer(mean_squared_error, squared=False),
}
# scaler = StandardScaler()
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(df[[i for i in df.columns if "D" in i]])

models = [
    LinearRegression(),
    KNeighborsRegressor(),
    RandomForestRegressor(),
    GradientBoostingRegressor(),
    AdaBoostRegressor(),
]
results = []

for model in models:
    cv_results = cross_validate(
        model, X_train_scaled, y_train, scoring=metrics_reg, return_train_score=True
    )
    cv_results = (
        pd.DataFrame(cv_results).drop(["fit_time", "score_time"], axis=1).mean()
    )
    cv_results["model"] = str(model)
    results.append(cv_results)
    print(str(model))
    print()

results_df = pd.concat(results, axis=1).T
```

```python
from sklearn.model_selection import KFold
from xgboost import XGBRegressor, XGBRFRegressor

kf = KFold(n_splits=5, shuffle=True, random_state=42)

results = []

for train_index, test_index in kf.split(X_train_scaled):
    X_train_kf, X_test_kf = X_train_scaled[train_index], X_train_scaled[test_index]
    y_train_kf, y_test_kf = y_train[train_index], y_train[test_index]
    model = XGBRFRegressor(objective="reg:squaredlogerror")
    model.fit(X_train_kf, y_train_kf)
    y_pred = model.predict(X_test_kf)

    train_MSE = mean_squared_error(y_train_kf, model.predict(X_train_kf))
    test_MSE = mean_squared_error(y_test_kf, y_pred)
    train_MAE = mean_absolute_error(y_train_kf, model.predict(X_train_kf))
    test_MAE = mean_absolute_error(y_test_kf, y_pred)
    train_MAX_ERROR = max_error(y_train_kf, model.predict(X_train_kf))
    test_MAX_ERROR = max_error(y_test_kf, y_pred)
    train_EV = explained_variance_score(y_train_kf, model.predict(X_train_kf))
    test_EV = explained_variance_score(y_test_kf, y_pred)
    train_RMSE = mean_squared_error(
        y_train_kf, model.predict(X_train_kf), squared=False
    )
    test_RMSE = mean_squared_error(y_test_kf, y_pred, squared=False)

    result_dict = {
        "train_MSE": train_MSE,
        "test_MSE": test_MSE,
        "train_MAE": train_MAE,
        "test_MAE": test_MAE,
        "train_MAX_ERROR": train_MAX_ERROR,
        "test_MAX_ERROR": test_MAX_ERROR,
        "train_EV": train_EV,
        "test_EV": test_EV,
        "train_RMSE": train_RMSE,
        "test_RMSE": test_RMSE,
    }

    results.append(result_dict)

xgb_df = pd.DataFrame(results)
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.model_selection import train_test_split

X_train_sgd, X_test_sgd, y_train_sgd, y_test_sgd = train_test_split(
    X_train_scaled, y_train, test_size=0.2, random_state=42
)

sgd_reg = SGDRegressor(max_iter=1000, tol=1e-3, eta0=1.0)

sgd_reg.fit(X_train_sgd, y_train_sgd)

y_pred_sgd = sgd_reg.predict(X_test_sgd)

results_dict = {
    "train_MSE": mean_squared_error(y_train_sgd, sgd_reg.predict(X_train_sgd)),
    "test_MSE": mean_squared_error(y_test_sgd, y_pred_sgd),
    "train_MAE": mean_absolute_error(y_train_sgd, sgd_reg.predict(X_train_sgd)),
    "test_MAE": mean_absolute_error(y_test_sgd, y_pred_sgd),
    "train_MAX_ERROR": max_error(y_train_sgd, sgd_reg.predict(X_train_sgd)),
    "test_MAX_ERROR": max_error(y_test_sgd, y_pred_sgd),
    "train_EV": explained_variance_score(y_train_sgd, sgd_reg.predict(X_train_sgd)),
    "test_EV": explained_variance_score(y_test_sgd, y_pred_sgd),
    "train_RMSE": mean_squared_error(
        y_train_sgd, sgd_reg.predict(X_train_sgd), squared=False
    ),
    "test_RMSE": mean_squared_error(y_test_sgd, y_pred_sgd, squared=False),
}

sgd_df = pd.DataFrame(results_dict, index=["SGD"])
sgd_df
```

```python
from sklearn.metrics import (
    mean_squared_error,
    make_scorer,
    mean_absolute_error,
    median_absolute_error,
    explained_variance_score,
    max_error,
)
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor

metrics_reg = {
    "MSE": make_scorer(mean_squared_error),
    "MAE": make_scorer(mean_absolute_error),
    "MAX_ERROR": make_scorer(max_error),
    "EV": make_scorer(explained_variance_score),
    "RMSE": make_scorer(mean_squared_error, squared=False),
}

kfold = KFold(n_splits=5, shuffle=True, random_state=42)

results = []

for train_idx, test_idx in kfold.split(X_train):
    X_train_kf, X_test_kf = X_train[train_idx], X_train[test_idx]
    y_train_kf, y_test_kf = y_train[train_idx], y_train[test_idx]

    scaler = MinMaxScaler()

    X_train_kf_scaled = scaler.fit_transform(X_train_kf)
    X_test_kf_scaled = scaler.transform(X_test_kf)

    model = RandomForestRegressor(
        bootstrap=True, max_samples=0.7, n_estimators=200, max_depth=10
    )
    model.fit(X_train_kf_scaled, y_train_kf)
    y_pred = model.predict(X_test_kf_scaled)

    result_dict = {
        "train_MSE": mean_squared_error(y_train_kf, y_train_kf),
        "test_MSE": mean_squared_error(y_test_kf, y_pred),
        "train_MAE": mean_absolute_error(y_train_kf, y_train_kf),
        "test_MAE": mean_absolute_error(y_test_kf, y_pred),
        "train_MAX_ERROR": max_error(y_train_kf, y_train_kf),
        "test_MAX_ERROR": max_error(y_test_kf, y_pred),
        "train_EV": explained_variance_score(y_train_kf, y_train_kf),
        "test_EV": explained_variance_score(y_test_kf, y_pred),
        "train_RMSE": mean_squared_error(y_train_kf, y_train_kf, squared=False),
        "test_RMSE": mean_squared_error(y_test_kf, y_pred, squared=False),
    }

another_df = pd.DataFrame(results)
another_df.mean()
```
