# following cell takes aprox 40 min to run

```python
from tqdm.notebook import tqdm
from sklearn.preprocessing import (
    MinMaxScaler,
    StandardScaler,
    PowerTransformer,
    Normalizer,
)
from sklearn.metrics import mean_squared_error, explained_variance_score
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import make_pipeline
from sklearn.decomposition import PCA

models = [
    LinearRegression(),
    DecisionTreeRegressor(),
    RandomForestRegressor(),
    SVR(),
    KNeighborsRegressor(),
]
pipe = make_pipeline(MinMaxScaler(), PCA(0.95))
scalers = [MinMaxScaler(), StandardScaler(), PowerTransformer(), Normalizer()]
results = []

for scaler in tqdm(scalers):
    result = {}
    pipe.steps[0] = (scaler.__class__.__name__, scaler)
    X_train_scaled = pipe.fit_transform(X_train_simpcv)
    X_val_scaled = pipe.transform(X_val_simpcv)

    for model in models:
        model.fit(X_train_scaled, y_train_simpcv)
        y_train_pred = model.predict(X_train_scaled)
        y_pred = model.predict(X_val_scaled)
        result = {
            "model": model.__class__.__name__,
            "scaler": scaler.__class__.__name__,
            "train_mse": mean_squared_error(y_train_pred, y_train_simpcv),
            "train_evs": explained_variance_score(y_train_pred, y_train_simpcv),
            "train_rmse": np.sqrt(mean_squared_error(y_train_pred, y_train_simpcv)),
            "test_mse": mean_squared_error(y_val_simpcv, y_pred),
            "test_evs": explained_variance_score(y_val_simpcv, y_pred),
            "test_rmse": np.sqrt(mean_squared_error(y_val_simpcv, y_pred)),
        }
        results.append(result)


results_df = pd.DataFrame(results)
results_df.to_csv("../data/results.csv", index=False)
```

## Decision - PCA with linear kernel aka regular PCA

takes 50 min to run on codespaces

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_validate
from sklearn.metrics import (
    mean_squared_error,
    explained_variance_score,
    max_error,
    make_scorer,
    mean_absolute_error,
)

pipe = make_pipeline(MinMaxScaler(), PCA(0.95))

# test various regression models
models = [
    LinearRegression(),
    DecisionTreeRegressor(),
    RandomForestRegressor(),
    SVR(),
    Lasso(),
    Ridge(),
]

# scoring metrics
metrics_reg = {
    "MSE": make_scorer(mean_squared_error),
    "MAE": make_scorer(mean_absolute_error),
    "MAX_ERROR": make_scorer(max_error),
    "EV": make_scorer(explained_variance_score),
    "RMSE": make_scorer(mean_squared_error, squared=False),
}

from sklearn.model_selection import KFold

# Define a KFold object
kf = KFold(n_splits=5, shuffle=True, random_state=42)

results = []

for model in models:
    # Create a new pipeline for each model
    pipe = make_pipeline(MinMaxScaler(), PCA(0.95))
    model_results = {}

    for train_index, test_index in kf.split(X_train):
        model_results = {}

        X_train_kf, X_test_kf = X_train[train_index], X_train[test_index]
        y_train_kf, y_test_kf = y_train[train_index], y_train[test_index]

        # scaling the data for each fold
        X_train_kf_scaled = pipe.fit_transform(X_train_kf)
        X_test_kf_scaled = pipe.transform(X_test_kf)

        # fit the model
        model.fit(X_train_kf_scaled, y_train_kf)

        # predict on the test set
        y_pred = model.predict(X_test_kf_scaled)
        y_train_pred = model.predict(X_train_kf_scaled)

        # Calculate metrics
        model_results["train_MSE"] = mean_squared_error(y_train_kf, y_train_pred)
        model_results["train_MAE"] = mean_absolute_error(y_train_kf, y_train_pred)
        model_results["train_MAX_ERROR"] = max_error(y_train_kf, y_train_pred)
        model_results["train_EV"] = explained_variance_score(y_train_kf, y_train_pred)
        model_results["train_RMSE"] = mean_squared_error(
            y_train_kf, y_train_pred, squared=False
        )

        model_results["test_MSE"] = mean_squared_error(y_test_kf, y_pred)
        model_results["test_MAE"] = mean_absolute_error(y_test_kf, y_pred)
        model_results["test_MAX_ERROR"] = max_error(y_test_kf, y_pred)
        model_results["test_EV"] = explained_variance_score(y_test_kf, y_pred)
        model_results["test_RMSE"] = mean_squared_error(
            y_test_kf, y_pred, squared=False
        )
        model_results["model"] = model.__class__.__name__
        results.append(model_results)
```

```python
# Utils for the notebook
from sklearn.metrics import (
    mean_squared_error,
    explained_variance_score,
    max_error,
    make_scorer,
    mean_absolute_error,
)
from sklearn.model_selection import cross_validate

# scoring metrics
metrics_reg = {
    "MSE": make_scorer(mean_squared_error),
    "MAE": make_scorer(mean_absolute_error),
    "MAX_ERROR": make_scorer(max_error),
    "EV": make_scorer(explained_variance_score),
    "RMSE": make_scorer(mean_squared_error, squared=False),
}


# cross validation function
def cv_results(model, X_train, y_train, metrics, hyperparam_name, hyperparams, cv=5):
    results = []
    # using cross_validate, we can split the data into 5 folds and get the training and validation scores for the metrics
    # across each fold iteration
    for hyperparam in hyperparams:
        model.set_params(**{hyperparam_name: hyperparam})
        cv_results = cross_validate(
            model, X_train, y_train, cv=cv, scoring=metrics, return_train_score=True
        )

        # mean of the k folds for each metric
        df = pd.DataFrame(cv_results).drop(["fit_time", "score_time"], axis=1).mean()
        results.append(df)
    results_df = pd.concat(results, axis=1).T
    results_df[hyperparam_name] = hyperparams

    return results_df
```
