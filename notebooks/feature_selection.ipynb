# Scaling

- MinMaxScaler
- StandardScaler
- PowerTransformer
- Normalizer

```python
from sklearn.preprocessing import (
    MinMaxScaler,
    StandardScaler,
    PowerTransformer,
    Normalizer,
)


mm_scaler = MinMaxScaler()
X_train_mm = mm_scaler.fit_transform(X_train)


X_ivs_mm = mm_scaler.transform(X_ivs)

s_scaler = StandardScaler()
X_train_sc = s_scaler.fit_transform(X_train)
X_ivs_sc = s_scaler.transform(X_ivs)

pt_scaler = PowerTransformer()
X_train_pt = pt_scaler.fit_transform(X_train)
X_ivs_pt = pt_scaler.transform(X_ivs)

n_scaler = Normalizer()
X_train_n = n_scaler.fit_transform(X_train)
X_ivs_n = n_scaler.transform(X_ivs)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.95)
pca.fit(X_train_mm)

X_train_pca = pca.fit_transform(X_train_mm)
X_ivs_pca = pca.transform(X_ivs_mm)

X_train_pca.shape
```

```python
pca = PCA(n_components=0.95)

X_train_pca = pca.fit_transform(X_train_sc)
X_ivs_pca = pca.transform(X_ivs_sc)

X_train_pca.shape
```

```python
pca = PCA(n_components=0.95)

X_train_pca = pca.fit_transform(X_train_pt)
X_ivs_pca = pca.transform(X_ivs_pt)

X_train_pca.shape
```

```python
pca = PCA(n_components=0.95)

X_train_pca = pca.fit_transform(X_train_n)
X_ivs_pca = pca.transform(X_ivs_n)

X_train_pca.shape
```

```python
df_ = pd.DataFrame(
    X_train_pca, columns=[f"PC{i}" for i in range(1, pca.n_components_ + 1)]
)
df_["y"] = y_train
```

```python
df_.corr()["y"].sort_values(ascending=False).head(20)
```

```python
from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    explained_variance_score,
    make_scorer,
    max_error,
)
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_validate

metrics_reg = {
    "MSE": make_scorer(mean_squared_error),
    "MAE": make_scorer(mean_absolute_error),
    "MAX_ERROR": make_scorer(max_error),
    "EV": make_scorer(explained_variance_score),
    "RMSE": make_scorer(mean_squared_error, squared=False),
}

lr = LinearRegression()

cv_results = cross_validate(
    lr, X_train_pca, y_train, scoring=metrics_reg, cv=5, return_train_score=True
)

pd.DataFrame(cv_results).drop(columns=["fit_time", "score_time"]).mean()
```

```python
from sklearn.svm import SVR

cv_results = cross_validate(
    SVR(),
    X_train_pca,
    y_train,
    scoring=metrics_reg,
    cv=5,
    return_train_score=True,
)

pd.DataFrame(cv_results).drop(columns=["fit_time", "score_time"]).mean()
```

```python
from sklearn.tree import DecisionTreeRegressor

cv_results = cross_validate(
    DecisionTreeRegressor(),
    X_train_pca,
    y_train,
    scoring=metrics_reg,
    cv=5,
    return_train_score=True,
)

pd.DataFrame(cv_results).drop(columns=["fit_time", "score_time"]).mean()
```

```python
from sklearn.ensemble import RandomForestRegressor

cv_results = cross_validate(
    RandomForestRegressor(),
    X_train_pca,
    y_train,
    scoring=metrics_reg,
    cv=5,
    return_train_score=True,
)

pd.DataFrame(cv_results).drop(columns=["fit_time", "score_time"]).mean()
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error, explained_variance_score
from sklearn.model_selection import train_test_split
from copy import deepcopy

X_train_, X_valid_, y_train_, y_valid_ = train_test_split(
    X_train_pca, y_train, test_size=0.2, random_state=42
)


sgd_reg = SGDRegressor(penalty=None, eta0=0.2, random_state=30)
n_epochs = 1000
best_valid_rmse = float("inf")
train_errors, val_errors, train_exp_vars, test_exp_vars = [], [], [], []


for epoch in range(n_epochs):
    sgd_reg.partial_fit(X_train_, y_train_)
    y_valid_predict = sgd_reg.predict(X_valid_)
    val_error = mean_squared_error(y_valid_, y_valid_predict, squared=False)
    exp_var = explained_variance_score(y_valid_, y_valid_predict)
    if val_error < best_valid_rmse:
        best_valid_rmse = val_error
        best_model = deepcopy(sgd_reg)

    # extra code  we evaluate the train error and save it for the figure
    y_train_predict = sgd_reg.predict(X_train_)
    train_error = mean_squared_error(y_train_, y_train_predict, squared=False)
    val_errors.append(val_error)
    train_errors.append(train_error)
    test_exp_vars.append(exp_var)
    train_exp_vars.append(explained_variance_score(y_train_, y_train_predict))
```

```python
fig, ax = plt.subplots(figsize=(10, 7), ncols=2, nrows=1)

plt.sca(ax[0])
plt.plot(val_errors, "b-", linewidth=3, label="train")
plt.plot(train_errors, "r--", linewidth=2, label="val")
plt.legend(loc="upper right", fontsize=14)
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("RMSE", fontsize=14)

plt.sca(ax[1])
plt.plot(train_exp_vars, "b-", linewidth=3, label="train")
plt.plot(test_exp_vars, "r--", linewidth=2, label="val")
plt.legend(loc="upper right", fontsize=14)
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("Explained Variance", fontsize=14)
```

```python
from sklearn.linear_model import lasso_path

eps = 5e-3  # the smaller it is the longer is the path

alphas_lasso, coefs_lasso, _ = lasso_path(X_train_pca, y_train, eps=eps)
neg_log_alphas = -np.log10(alphas_lasso)
```

```python
plt.plot(neg_log_alphas, coefs_lasso.T, linewidth=2)
plt.xlabel("-Log(alpha)")
plt.ylabel("coefficients")
plt.title("Lasso Path")
plt.show()
```
