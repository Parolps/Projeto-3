```python
from tqdm.notebook import tqdm
from sklearn.preprocessing import (
    MinMaxScaler,
    StandardScaler,
    PowerTransformer,
    Normalizer,
)
from sklearn.metrics import mean_squared_error, explained_variance_score
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.compose import ColumnTransformer

# from sklearn.decomposition import PCA     # with/without pca

models = [
    LinearRegression(),
    Ridge(),
    RandomForestRegressor(),
    SVR(),
    KNeighborsRegressor(),
]

scalers = [MinMaxScaler(), StandardScaler(), Normalizer()]
results = []

FP_cols = [col for col in df.columns if "FP" in col]
D_cols = [col for col in df.columns if "D" in col]

ct = ColumnTransformer(
    [("scaler", MinMaxScaler(), D_cols)],
    remainder="passthrough",
    verbose_feature_names_out=True,
)


for scaler in tqdm(scalers):
    ct.set_params(scaler=scaler)
    X_train_sc = ct.fit_transform(X_train_simpcv)
    X_val_sc = ct.transform(X_val_simpcv)

    for model in models:
        model.fit(X_train_sc, y_train_simpcv)
        y_pred = model.predict(X_val_sc)
        y_train_pred = model.predict(X_train_sc)
        result = {
            "model": model.__class__.__name__,
            "scaler": scaler.__class__.__name__,
            "train_mse": mean_squared_error(y_train_pred, y_train_simpcv),
            "test_mse": mean_squared_error(y_val_simpcv, y_pred),
            "train_rmse": np.sqrt(mean_squared_error(y_train_pred, y_train_simpcv)),
            "test_rmse": np.sqrt(mean_squared_error(y_val_simpcv, y_pred)),
            "train_evs": explained_variance_score(y_train_pred, y_train_simpcv),
            "test_evs": explained_variance_score(y_val_simpcv, y_pred),
        }
        results.append(result)

results_df = pd.DataFrame(results)
results_df.to_csv("../data/results.csv")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import make_scorer, mean_squared_error, explained_variance_score
from sklearn.svm import SVR
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
import numpy as np
import pandas as pd

# Initialize SVR
svr = SVR(kernel="linear")

# Initialize lists to store scores
explained_variance_scores = []
rmse_scores = []

# For each number of features from 1050 to 1
for num_features in range(1050, 0, -1):
    # Select the top num_features
    top_features = (
        feature_importances_df.iloc[:num_features]
        .index.str.replace("PC", "")
        .astype(int)
        - 1
    )

    # Perform cross-validation
    cv_scores_variance = []
    cv_scores_rmse = []
    for train_index, val_index in KFold(n_splits=5).split(X_train):
        # Split the data
        X_train_fold, X_val_fold = X_train[train_index], X_train[val_index]
        y_train_fold, y_val_fold = y_train[train_index], y_train[val_index]

        # get the columns name fot the ct to work
        X_train_fold = pd.DataFrame(X_train_fold, columns=col_names)
        X_val_fold = pd.DataFrame(X_val_fold, columns=col_names)

        # Fit and transform the data with the pipeline
        pipe = Pipeline(
            [
                ("ct", ct),
                ("pca", PCA(n_components=1050)),
            ]
        )
        X_train_transformed = pipe.fit_transform(X_train_fold)
        X_val_transformed = pipe.transform(X_val_fold)

        # Select the top features
        X_train_selected = X_train_transformed[:, top_features]
        X_val_selected = X_val_transformed[:, top_features]

        # Fit the model and compute scores
        svr.fit(X_train_selected, y_train_fold)
        y_pred = svr.predict(X_val_selected)
        cv_scores_variance.append(explained_variance_score(y_val_fold, y_pred))
        cv_scores_rmse.append(np.sqrt(mean_squared_error(y_val_fold, y_pred)))

    # Compute the mean scores
    explained_variance = np.mean(cv_scores_variance)
    rmse = np.mean(cv_scores_rmse)

    # Append scores to the lists
    explained_variance_scores.append(explained_variance)
    rmse_scores.append(rmse)
```

## Decision - PCA with linear kernel aka regular PCA

takes 50 min to run on codespaces

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_validate
from sklearn.metrics import (
    mean_squared_error,
    explained_variance_score,
    max_error,
    make_scorer,
    mean_absolute_error,
)

pipe = make_pipeline(MinMaxScaler(), PCA(0.95))

# test various regression models
models = [
    LinearRegression(),
    DecisionTreeRegressor(),
    RandomForestRegressor(),
    SVR(),
    Lasso(),
    Ridge(),
]

# scoring metrics
metrics_reg = {
    "MSE": make_scorer(mean_squared_error),
    "MAE": make_scorer(mean_absolute_error),
    "MAX_ERROR": make_scorer(max_error),
    "EV": make_scorer(explained_variance_score),
    "RMSE": make_scorer(mean_squared_error, squared=False),
}

from sklearn.model_selection import KFold

# Define a KFold object
kf = KFold(n_splits=5, shuffle=True, random_state=42)

results = []

for model in models:
    # Create a new pipeline for each model
    pipe = make_pipeline(MinMaxScaler(), PCA(0.95))
    model_results = {}

    for train_index, test_index in kf.split(X_train):
        model_results = {}

        X_train_kf, X_test_kf = X_train[train_index], X_train[test_index]
        y_train_kf, y_test_kf = y_train[train_index], y_train[test_index]

        # scaling the data for each fold
        X_train_kf_scaled = pipe.fit_transform(X_train_kf)
        X_test_kf_scaled = pipe.transform(X_test_kf)

        # fit the model
        model.fit(X_train_kf_scaled, y_train_kf)

        # predict on the test set
        y_pred = model.predict(X_test_kf_scaled)
        y_train_pred = model.predict(X_train_kf_scaled)

        # Calculate metrics
        model_results["train_MSE"] = mean_squared_error(y_train_kf, y_train_pred)
        model_results["train_MAE"] = mean_absolute_error(y_train_kf, y_train_pred)
        model_results["train_MAX_ERROR"] = max_error(y_train_kf, y_train_pred)
        model_results["train_EV"] = explained_variance_score(y_train_kf, y_train_pred)
        model_results["train_RMSE"] = mean_squared_error(
            y_train_kf, y_train_pred, squared=False
        )

        model_results["test_MSE"] = mean_squared_error(y_test_kf, y_pred)
        model_results["test_MAE"] = mean_absolute_error(y_test_kf, y_pred)
        model_results["test_MAX_ERROR"] = max_error(y_test_kf, y_pred)
        model_results["test_EV"] = explained_variance_score(y_test_kf, y_pred)
        model_results["test_RMSE"] = mean_squared_error(
            y_test_kf, y_pred, squared=False
        )
        model_results["model"] = model.__class__.__name__
        results.append(model_results)
```
