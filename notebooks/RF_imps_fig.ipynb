```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.options.display.max_columns = None
plt.style.use("fivethirtyeight")
plt.rcParams["figure.dpi"] = 150
```

```python
importances = pd.read_csv("../data/feature_importances.csv")
importances.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
top_idx = importances.mean().sort_values(ascending=False).head(20).index
sns.boxplot(
    importances.loc[:, top_idx],
    linewidth=0.9,
    color="#FF6C00",
    linecolor="k",
    fliersize=2,
)
# plt.xlabel("Feature Rank", fontsize=14)
plt.xticks(
    ticks=range(0, len(top_idx)), labels=[i for i in top_idx], fontsize=8, rotation=45
)
plt.ylabel("Feature Importance", fontsize=14)
plt.yticks(fontsize=8)
plt.savefig("../figs/feature_importances.png", bbox_inches="tight")
plt.show()
```
