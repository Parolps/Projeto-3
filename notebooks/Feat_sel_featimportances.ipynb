```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import pickle
import seaborn as sns

pd.options.display.max_columns = None
plt.style.use("fivethirtyeight")
plt.rcParams["figure.dpi"] = 150
```

```python
X_train, X_ivs, y_train, col_names = pickle.load(open("../data/drd2_data.pickle", "rb"))
```

```python
df = pd.DataFrame(X_train, columns=col_names)
df.head()
```

```python
cols = [i for i in df.columns if "F" in i]
FPs = df[cols].copy()
```

## Using random forests to do feature selection

Takes 22 min to run on codespaces

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100)
rf.fit(FPs, y_train)

importances = pd.DataFrame()

for i in range(0, 11):
    rf = RandomForestRegressor(n_estimators=100)
    rf.fit(FPs, y_train)
    importances[f"{i}"] = rf.feature_importances_
```

```python
importances.index = cols
```

```python
importances.mean(axis=1).sort_values(ascending=False)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import SelectFromModel

sel = SelectFromModel(RandomForestRegressor(n_estimators=100))
sel.fit(FPs, y_train)

sel.get_support()

selected_feat = FPs.columns[(sel.get_support())]
print(len(selected_feat))
print(selected_feat)
```
